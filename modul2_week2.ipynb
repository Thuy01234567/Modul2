```python
import numpy as np
def compute_vector_length(vector):
  # *********** Your code here ***********
  len_of_vector = np.linalg.norm(vector)
  return len_of_vector
vector = np.array([-2, 4, 9, 21])
result = compute_vector_length([vector])
print (round(result,2))
```

```python
def compute_dot_product( vector1,vector2 ):
  # *********** Your code here ***********
  result = np.dot(vector1,vector2)
  return result
v1 = np.array([0, 1, -1, 2])
v2 = np.array([2, 5, 1, 0])
result = compute_dot_product (v1 , v2)
print (round(result,2) )
```

```python
import numpy as np
def matrix_multi_vector( matrix, vector ):
  # *********** Your code here ***********
  result = np.dot(matrix, vector)
  return result
m = np.array ([[ -1, 1, 1], [0, -4, 9]])
v = np.array ([0, 2, 1])
result = matrix_multi_vector(m, v)
print ( result )
```

```python
import numpy as np
def matrix_multi_matrix(matrix1, matrix2 ):
  # *********** Your code here ***********
  result = np.dot(matrix1,matrix2)
  return result
m1 = np.array ([[0, 1, 2], [2, -3, 1]])
m2 = np.array ([[1, -3],[6, 1], [0, -1]])
result = matrix_multi_matrix (m1, m2)
print(result )
```

```python
import numpy as np
def inverse_matrix( matrix ):
  # *********** Your code here ***********
  result = np.linalg.inv(matrix)
  return result
m1 = np.array([[-2, 6], [8, -4]])
result = inverse_matrix(m1)
print(result)
```

```python
def compute_eigenvalues_eigenvectors(matrix ):
  # *********** Your code here ***********
  eigenvalues, eigenvectors = np.linalg.eig(matrix)
  return eigenvalues, eigenvectors
matrix = np.array ([[0.9, 0.2], [0.1, 0.8]])
eigenvalues, eigenvectors = compute_eigenvalues_eigenvectors(matrix)
print(eigenvectors)
```

```python
def compute_cosine (v1 , v2) :
  # *********** Your code here ***********
  cos_sim = np.dot(v1,v2)/(np.linalg.norm(v1)*np.linalg.norm(v2))
  return cos_sim
x = np.array ([1, 2, 3, 4])
y = np.array ([1, 0, 3, 0])
result = compute_cosine (x,y)
print(round(result, 3) )
```

```python
#Resize các ảnh đầu vào về cùng kích thước:
import numpy as np
from google.colab.patches import cv2_imshow
import cv2

bg1_image = cv2.imread('/content/GreenBackground.png', 1)
bg1_image = cv2.resize(bg1_image, (678, 381))

ob_image = cv2.imread('/content/Object.png', 1)
ob_image = cv2.resize(ob_image, (678, 381) )

bg2_image = cv2.imread('/content/NewBackground.jpg', 1)
bg2_image = cv2.resize(bg2_image, (678, 381))

#Xây dựng hàm compute_difference():
def compute_difference(bg_img, input_img):
  # *********** Your code here ***********
  difference_single_channel = cv2.absdiff(bg_img, input_img)
  return difference_single_channel
difference_single_channel = compute_difference(bg1_image, ob_image)
cv2_imshow(difference_single_channel)

# Xây dựng hàm compute_binary_mask():
def compute_binary_mask(difference_single_channel):
  # *********** Your code here ***********
  _, difference_binary = cv2.threshold(difference_single_channel, 15, 255, cv2.THRESH_BINARY) # Extract
  return difference_binary
binary_mask = compute_binary_mask(difference_single_channel)
cv2_imshow(binary_mask)

def replace_background (bg1_image, bg2_image, ob_image):
  # *********** Your code here ***********
  difference_single_channel = compute_difference(bg1_image, ob_image)
  binary_mask = compute_binary_mask(difference_single_channel)
  output = np.where(binary_mask ==255, ob_image, bg2_image)
  return output
output = replace_background(bg1_image, bg2_image, ob_image)
cv2_imshow(output)
```
